# Simple Line Plots

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

#### We start by creating a figure and an axes. In their simplest form:

```python
fig = plt.figure()
ax = plt.axes()
```

- **``plt.Figure``**: a single container that contains all the objects representing axes, graphics, text, and labels.
- **``plt.Axes``**: a bounding box with ticks and labels, which will eventually contain the plot elements that make up our visualization.
- **Convention:**
    - **``fig``**: a figure instance
    - **``ax``**: an axes instance or group of axes instances.

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x));
```

Alternatively, we can use the **Matlab interface**:

```python
plt.plot(x, np.sin(x));
```

**Single figure with multiple lines:** call the ``plot`` function multiple times:

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

## Adjusting the Plot: Line Colors and Styles

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

**``linestyle``** and **``color``**: combined into a single non-keyword argument

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

## Adjusting the Plot: Axes Limits
**Beyond default axes limits: ``plt.xlim()`` & ``plt.ylim()``**

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

**Reverse display of axis: reverse the order of the arguments**

```python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2);
```

**``plt.axis()``**: set the **``x``** & **``y``** limits with a single call
- params list: **``[xmin, xmax, ymin, ymax]``**

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5]);
```

**Automatically tighten the bounds around the current plot**

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

**Ensure an equal aspect ratio so that on your screen, one unit in ``x`` is equal to one unit in ``y``:

```python
plt.plot(x, np.sin(x))
plt.axis('equal');
```

## Labeling Plots: titles, axis labels, and simple legends

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

## Aside: Matplotlib Gotchas
For transitioning between MATLAB-style functions and object-oriented methods, make the following changes:

- ``plt.xlabel()``   ``ax.set_xlabel()``
- ``plt.ylabel()``  ``ax.set_ylabel()``
- ``plt.xlim()``   ``ax.set_xlim()``
- ``plt.ylim()``  ``ax.set_ylim()``
- ``plt.title()``  ``ax.set_title()``

In the object-oriented interface to plotting, use the ``ax.set()`` method to set all these properties at once:

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
```
